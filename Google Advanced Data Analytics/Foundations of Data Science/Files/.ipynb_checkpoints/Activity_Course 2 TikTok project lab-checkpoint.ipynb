A continuación, examina las tendencias de interacción asociadas con cada estado de reclamación.

Comienza usando el enmascaramiento booleano para filtrar los datos según el estado de la reclamación y, a continuación, calcula la media y la mediana de visualizaciones para cada estado.

```python
# What is the average view count of videos with "claim" status?

# Filtrar por categoría usando enmascaramiento booleano
mask_A = data["claim_status"] == "claim"


# Calcular media y mediana para cada categoría
print("claim:")
print("Media:", data.loc[mask_A, "video_view_count"].mean())
print("Mediana:", data.loc[mask_A, "video_view_count"].median())

print("Media:", data.loc[mask_A, "video_like_count"].mean())
print("Mediana:", data.loc[mask_A, "video_like_count"].median())

print("Media:", data.loc[mask_A, "video_share_count"].mean())
print("Mediana:", data.loc[mask_A, "video_share_count"].median())

print("Media:", data.loc[mask_A, "video_download_count"].mean())
print("Mediana:", data.loc[mask_A, "video_download_count"].median())

print("Media:", data.loc[mask_A, "video_comment_count"].mean())
print("Mediana:", data.loc[mask_A, "video_comment_count"].median())
```

```python
# What is the average view count of videos with "opinion" status?
mask_B = data["claim_status"] == "opinion"

print("\nopinion:")
print("Media:", data.loc[mask_B, "video_view_count"].mean())
print("Mediana:", data.loc[mask_B, "video_view_count"].median())

print("Media:", data.loc[mask_B, "video_like_count"].mean())
print("Mediana:", data.loc[mask_B, "video_like_count"].median())

print("Media:", data.loc[mask_B, "video_share_count"].mean())
print("Mediana:", data.loc[mask_B, "video_share_count"].median())

print("Media:", data.loc[mask_B, "video_download_count"].mean())
print("Mediana:", data.loc[mask_B, "video_download_count"].median())

print("Media:", data.loc[mask_B, "video_comment_count"].mean())
print("Mediana:", data.loc[mask_B, "video_comment_count"].median())
```

**Pregunta:** ¿Qué observas sobre la media y los medios dentro de cada categoría de reclamación?

Ahora, examina las tendencias asociadas con el estado de prohibición del autor.

Usa `groupby()` para calcular cuántos vídeos hay para cada combinación de categorías de estado de reclamación y estado de prohibición del autor.

```python
# Get counts for each group combination of claim status and author ban status
# Agrupar por ambas columnas y contar
conteos = data.groupby(["claim_status","author_ban_status"]).size().reset_index(name="frecuencia")

print(conteos)
```

**Pregunta:** ¿Qué observas sobre la cantidad de videos con reclamos de autores baneados? ¿Por qué podría darse esta relación?

Continúa investigando los niveles de interacción, centrándote ahora en `author_ban_status`.

Calcula la mediana de compartición de videos para cada estado de baneo del autor. 

```python
### YOUR CODE HERE ###
```

```python
# What's the median video share count of each author ban status?
# Filtrar por categoría usando enmascaramiento booleano
mask_C = data["author_ban_status"] == "active"
mask_D = data["author_ban_status"] == "banned"
mask_E = data["author_ban_status"] == "under review"

print("Mediana:", data.loc[mask_C, "video_share_count"].median())
print("Mediana:", data.loc[mask_D, "video_share_count"].median())
print("Mediana:", data.loc[mask_E, "video_share_count"].median())
```

**Pregunta:** ¿Qué observas sobre el número de compartidos de los autores baneados, en comparación con el de los autores activos? Analiza esto con más detalle.

Usa `groupby()` para agrupar los datos por `author_ban_status` y, a continuación, usa `agg()` para obtener el número, la media y la mediana de cada una de las siguientes columnas:

* `video_view_count`
* `video_like_count`
* `video_share_count`

Recuerda que el argumento de la función ``agg()`` es un diccionario cuyas claves son columnas. Los valores de cada columna son una lista de los cálculos que quieres realizar.

```python
vistas = data.groupby("author_ban_status")["video_view_count"].agg(["mean", "median", "count"])
likes = data.groupby("author_ban_status")["video_like_count"].agg(["mean", "median", "count"])
compartidos = data.groupby("author_ban_status")["video_share_count"].agg(["mean", "median", "count"])

print(vistas)
print(likes)
print(compartidos)
```

**Pregunta:** ¿Qué observas sobre la cantidad de visualizaciones, "me gusta" y comparticiones de los autores bloqueados en comparación con los autores activos?

Ahora, crea tres nuevas columnas para comprender mejor las tasas de interacción:

`likes_per_view`: representa la cantidad de "me gusta" dividida entre la cantidad de visualizaciones de cada video.

`comments_per_view`: representa la cantidad de comentarios dividida entre la cantidad de visualizaciones de cada video.

`shares_per_view`: representa la cantidad de comparticiones dividida entre la cantidad de visualizaciones de cada video.

```python
# Create a likes_per_view column
data["likes_per_view"] = data["video_like_count"] / data["video_view_count"]

# Create a comments_per_view column
data["comments_per_view"] = data["comments_like_count"] / data["video_view_count"]

# Create a shares_per_view column
data["shares_per_view"] = data["shares_like_count"] / data["video_view_count"]
```

Utilice `groupby()` para compilar la información en cada una de las tres columnas recién creadas para cada combinación de categorías de estado de reclamo y estado de prohibición del autor, luego utilice `agg()` para calcular el recuento, la media y la mediana de cada grupo.

```python
resumen = (
    data.groupby(["claim_status", "author_ban_status"])
        .agg({
            "likes_per_view": ["count","mean","median"],
            "comments_per_view": ["count", "mean", "median"], 
            "shares_per_view": ["count", "mean", "median"]})
)

resumen = resumen.reset_index()

print(resumen)
```
